```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import os
lib=[]
lib_path=[]
def process_name(libname):
	if("#2e" in libname):
		libname_temp=libname.replace("#2e",".")
	elif("#2d" in libname):
		libname_temp=libname.replace("#2d","-")
	elif("#2b" in libname):
		libname_temp=libname.replace("#2b","+")
	else:
		libname_temp=libname
	return libname_temp
	
def process_include(line):
	line=line.rstrip()
	index=line.find("/")
	include_path=line[index:]
	return include_path
def process_define(line):
	line=line.rstrip()
	line=line.replace("DEFINE","")
	index_1=line.find("/")
	lib_path=line[index_1:]
	temp_line=line[:index_1]
	temp_line=temp_line.replace(" ","")
	lib_name=process_name(temp_line)

	return [lib_name,lib_path]
def read_cds(path):
	if(os.path.exists(path)):	
		with open (path,"r") as f:
			for l in f.readlines():
				if(("INCLUDE" in l or "include" in l) and (not l[0]=="#")):
					include_path=process_include(l)
					read_cds(include_path)
					#print(l.rstrip())
				elif("UNDEFINE" in l or "undefine" in l):
					pass
				elif(("DEFINE" in l or "define" in l) and (not l[0]=="#")):
					#print(l.rstrip())
					lib_list=process_define(l)
					#print(lib_list)
					lib.append(lib_list[0])
					lib_path.append(lib_list[1])

	else:
		print("{0} doesn't exist".format(path))
def get_cellview(path):
	dbtype_list_new=[]
	if(os.path.exists(path)):
		dbtype_list = os.listdir(path)
		#print(dbtype_list)
		for dbtype in dbtype_list:
			if(os.path.isdir(os.path.join(path,dbtype))):
				dbtype_list_new.append(process_name(dbtype))
	else:
		print("{0} doesn't exist".format(path))		
	return dbtype_list_new


read_cds("cds.lib")
lib_temp=[]
for c in lib:
	lib_temp.append(c)
lib.sort()
lib_path_temp=lib_path
lib_path=[]
len_lib=len(lib)
for i in range(0,len_lib):
	original_index=lib_temp.index(lib[i])
	lib_path.append(lib_path_temp[original_index])
#--------------------generate function----------------------
def generate_il(path,cell_list):
	text=open("./schematic.il","w+")
	envOp='''simSetEnvOptions() 
simEnvOptionsForm->simFormIncrNetlist->myChoice->value =  nil 
simEnvOptionsForm->simFormNetlistHier->myChoice->value =  t 
hiFormDone(simEnvOptionsForm) '''
	print(envOp,file=text)
	for c in cell_list:
		temp=c.split(":")
		lib=temp[0]
		cell=temp[1]
		if(path[-1]=="/"):
			path_module=path+c
		elif( not path[-1]=="/"):
			path_module=path+"/"+c
		script='''
simInitEnvWithArgs( 
"{0}"
"{1}"
"{2}"
"schematic"
"pscan2"
"forceInit"
) 

simRunNetAndSimWithArgs( 
"{3}"
"{4}"
"schematic"
"pscan2"
t
nil
nil
10
) '''.format(path_module,lib,cell,lib,cell)
		print(script,file=text)
	text.close()
def generate_schematic_file(path,cell_list):
	if(path[-1]=="/"):
		schematic_file=path+"schematic_file"
	else:
		schematic_file=path+"/schematic_file"
	text=open("{0}".format(schematic_file),"w+")
	for c in cell_list:
		print(c,file=text)
	text.close()
#-------------------------------------------------------------
from tkinter import *
import tkinter

Lb2_value_list=[[] for i in range(int(len_lib))]
for i in range(0,len_lib):
	cellview=get_cellview(lib_path[i])
	cellview.sort()
	Lb2_value_list[i]=cellview
	

top = tkinter.Tk()
#top.geometry("800x600")
top.title("Library Manager - Lite")

frame = Frame(top)
frame.pack()
def show_lb2(*args):
	global lib_name
	if(Lb1.curselection()==()):
		pass
	else:
		indexs = Lb1.curselection()
		index = int(indexs[0])
		lib_name=Lb1.get(Lb1.curselection())
		v.set(tuple(Lb2_value_list[index]))

def read_name(*args):
    global name
    if(Lb2.curselection()==()):
        pass
    else:
        select_list=list(Lb2.curselection())
        #print(select_list)
        len_sel_list=len(select_list)
        name=[]
        #print(Lb2.curselection())
        for i in range(0,len_sel_list):
            #print(Lb2.get(Lb2.curselection()))
            index=tuple((select_list[i],))
            name.append(Lb2.get((index)))
def read_lb3(*args):
    global remove_list
    if(Lb3.curselection()==()):
        pass
    else:
        rm_list=list(Lb3.curselection())
        len_rm_list=len(rm_list)
        remove_list=[]
        for i in range(0,len_rm_list):
            #print(Lb2.get(Lb2.curselection()))
            index=tuple((rm_list[i],))
            remove_list.append(Lb3.get((index)))

frame_2 = Frame(frame)# 第二层frame，右frame，长在主frame上
frame_2.pack(side='left')
frame_3 = Frame(frame)# 第二层frame，右frame，长在主frame上
frame_3.pack(side='left')
frame_4 = Frame(frame)# 第二层frame，右frame，长在主frame上
frame_4.pack(side='left')
frame_5 = Frame(frame)# 第二层frame，右frame，长在主frame上
frame_5.pack(side='left')
Lb1 = Listbox(frame_2,height=30,width=30)
Lb1.bind("<<ListboxSelect>>",show_lb2)
for i in range(0,len_lib):
	Lb1.insert(i,lib[i])
Lb1.pack()
v = StringVar()
Lb2 = Listbox(frame_3,listvariable=v,selectmode = EXTENDED,height=30,width=40)
Lb2.bind("<<ListboxSelect>>",read_name)
Lb2.pack()
sel_v= StringVar()
Lb3 = Listbox(frame_5,listvariable=sel_v,selectmode = EXTENDED,height=30,width=40)
Lb3.bind("<<ListboxSelect>>",read_lb3)
toLb3_list=[]
def add_value():
	for n in name:
		if( lib_name+":"+n in toLb3_list):
			pass
		else:
			toLb3_list.append(lib_name+":"+n)
	sel_v.set(tuple(toLb3_list))

def remove_cell():
	for r in remove_list:
		toLb3_list.remove(r)
	sel_v.set(tuple(toLb3_list))
add_cell = tkinter.Button(frame_4, text =" Add ", command = add_value)
add_cell.pack()
rm_cell = tkinter.Button(frame_4, text ="Delete", command = remove_cell)
rm_cell.pack()
Lb3.pack()

def TarDirget():
	global path_dir
	global cell_list_lb3
	path_dir=Tardir_entry.get()
	cell_list_lb3=toLb3_list
	generate_il(path_dir,cell_list_lb3)
	generate_schematic_file(path_dir,cell_list_lb3)
	top.destroy()
Tardir_label = Label(top, text="Target Dir")
Tardir_label.pack( side = LEFT)
entry_var = StringVar() 
Tardir_entry = Entry(top, textvariable=entry_var,bd =1,width=40)
Tardir_entry.pack(side = LEFT)
entry_var.set('~/Cadence/designs/') 
Generate_button = tkinter.Button(top, text ="Generate", command = TarDirget)
Generate_button.pack()
top.mainloop()
```

```python
import tkinter

win = tkinter.Tk()
win.title("Listbox列表框")
win.geometry("800x600+600+100")
'''
列表框控件，可以包含一个或者多个文本框
作用：在listbox控件的小窗口显示一个字符串
'''
#1.创建一个listbox,添加几个元素
lb=tkinter.Listbox(win,selectmode=tkinter.BROWSE)
lb.pack()
for item in["good","nice","handsome","very good","verynice"]:
    #按顺序添加(导入到列表)
    lb.insert(tkinter.END,item)
#在列表前添加
lb.insert(tkinter.ACTIVE,"cool")
#将列表当一个来添加
#lb.insert(tkinter.END,["very good","very nice"])

#删除
#参数1为开始的索引，参数2为结束的索引，如果只有参数1，则删除参数1为下标的值
#lb.delete(1,3)#删除1到3
#lb.delete(1)#只删除下标为1的数

#选中
#参数1为开始的索引，参数2为结束的索引，如果只有参数1，则选中参数1为下标的值
lb.select_set(2,4)#选中2到4
#lb.select_set(2)

#取消选中
#参数1为开始的索引，参数2为结束的索引，如果只有参数1，则取消选中参数1为下标的值
#lb.select_clear(2,4)
lb.select_clear(3)

#获取列表中元素的个数
print(lb.size())

#取值
#参数1为开始的索引，参数2为结束的索引，如果只有参数1，则只取参数1为下标的值
print(lb.get(2,4))
print(lb.get(2))

#返回当前选中的索引项，不是item元素
print(lb.curselection())#(2, 4)2和4被选中

#判断一个选项是否被选中(选中返回True,反之返回False)
print(lb.select_includes(1))#False
print(lb.select_includes(2))#True

win.mainloop()
```
