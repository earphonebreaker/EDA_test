```python
from pylab import *
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import random
import time
def test_random_block_detect(try_num):
    for attemp in range(try_num):
        #--------------------产生随机坐标-------------------------
        random_list=[]
        def random_group(num,x_min_value,x_max_value,y_min_value,y_max_value):
            for i in range(num):
                start=[[random.randint(x_min_value,x_max_value),random.randint(y_min_value,y_max_value)],random.randint(1,4)]
                end=[[random.randint(x_min_value,x_max_value),random.randint(y_min_value,y_max_value)],random.randint(1,4)]
                #print(start,end)
                random_list.append([end,start])
        #random_group(10,-100,0,-100,30)
        #random_group(50,60,150,60,210)
        #random_group(50,10,50,250,300)
        #random_group(50,30,80,-100,-50)
        #random_group(10,-100,100,-100,100)
        #random_group(10,100,300,100,300)
        random_group(10,-100,300,-100,300)

        #print(len(random_list))
        #for r in random_list:
        #    print(r)
        #random_list=[[[[-65, -92], 2], [[-78, -52], 1]], [[[-39, -83], 2], [[-4, -52], 4]]]
        #---------------------找到margin-----------------
        block_margin=1
        range_list=[]
        coord_list=[]
        start_time=time.time()
        first_start=random_list[0][0][0]
        first_end=random_list[0][1][0]
        first_start_x=first_start[0]
        first_start_y=first_start[1]
        first_end_x=first_end[0]
        first_end_y=first_end[1]
        new_range=[[min(first_start_x,first_end_x)-block_margin,min(first_start_y,first_end_y)-block_margin],\
                   [max(first_start_x,first_end_x)+block_margin,max(first_start_y,first_end_y)+block_margin]]
        range_list.append(new_range)
        coord_list.append([[first_start,first_end]])

        #print(range_list)
        for r in range(1,len(random_list)):
            start=random_list[r][0][0]
            end=random_list[r][1][0]
            start_x=start[0]
            start_y=start[1]
            end_x=end[0]
            end_y=end[1]

            overlap_list=[]
            for o in range(len(range_list)):
                range_x_min=range_list[o][0][0]
                range_x_max=range_list[o][1][0]
                range_y_min=range_list[o][0][1]
                range_y_max=range_list[o][1][1]
                mod_bottomleft=[min(start_x,end_x)-block_margin,min(start_y,end_y)-block_margin]
                mod_topright=[max(start_x,end_x)+block_margin,max(start_y,end_y)+block_margin]
                
                center_dx=abs((mod_bottomleft[0]+mod_topright[0])/2-(range_x_min+range_x_max)/2)
                center_dy=abs((mod_bottomleft[1]+mod_topright[1])/2-(range_y_min+range_y_max)/2)
                mod_width=abs(mod_topright[0]-mod_bottomleft[0])
                mod_height=abs(mod_topright[1]-mod_bottomleft[1])
                range_width=abs(range_x_max-range_x_min)
                range_height=abs(range_y_max-range_y_min)
                
                condition_loop=(center_dx<=(mod_width+range_width)/2)and(center_dy<=(mod_height+range_height)/2)

                
                if(condition_loop):
                    overlap_list.append(o)
                else:
                    pass
            #print(overlap_list)
            if(overlap_list==[]):#说明是新的范围
                new_range=[[min(start_x,end_x)-block_margin,min(start_y,end_y)-block_margin],\
                            [max(start_x,end_x)+block_margin,max(start_y,end_y)+block_margin]]
                range_list.append(new_range)
                coord_list.append([[start,end]])
            elif(len(overlap_list)==1):#只有一个重合了
                index=overlap_list[0]
                new_range=[[min(start_x,end_x)-block_margin,min(start_y,end_y)-block_margin],\
                            [max(start_x,end_x)+block_margin,max(start_y,end_y)+block_margin]]
                coord_list[index].append([start,end])

                range_value=range_list[index]

                range_xmin_list=[new_range[0][0],range_value[0][0]]
                range_xmax_list=[new_range[1][0],range_value[1][0]]
                range_ymin_list=[new_range[0][1],range_value[0][1]]
                range_ymax_list=[new_range[1][1],range_value[1][1]] 
                new_range_sum=[[min(range_xmin_list),min(range_ymin_list)],[max(range_xmax_list),max(range_ymax_list)]]
                range_list[index]=new_range_sum        

            else: #说明几个范围有重合，需要把他们融合在一起
                new_range=[[min(start_x,end_x)-block_margin,min(start_y,end_y)-block_margin],\
                            [max(start_x,end_x)+block_margin,max(start_y,end_y)+block_margin]]
                for i in range(1,len(overlap_list)):#把后面的元素添加到第一个list中
                    for coord in coord_list[overlap_list[i]]:
                        coord_list[overlap_list[0]].append(coord)
                range_xmin_list=[new_range[0][0]]
                range_xmax_list=[new_range[1][0]]
                range_ymin_list=[new_range[0][1]]
                range_ymax_list=[new_range[1][1]]
                for i in range(0,len(overlap_list)):
                    index=overlap_list[i]
                    range_value=range_list[index]
                    range_xmin_list.append(range_value[0][0])
                    range_xmax_list.append(range_value[1][0])
                    range_ymin_list.append(range_value[0][1])
                    range_ymax_list.append(range_value[1][1])
                #print(range_xmin_list)
                #print(range_xmax_list)
                #print(range_ymin_list)
                #print(range_ymax_list)
                new_range_sum=[[min(range_xmin_list),min(range_ymin_list)],[max(range_xmax_list),max(range_ymax_list)]]
                range_list[overlap_list[0]]=new_range_sum
                for i in range(1,len(overlap_list),-1):
                    index=overlap_list[i]
                    range_list.pop(index)
                    coord_list.pop(index)
            #print(range_list)

        #for r in range_list:
        #    print(r)
        
        
        len_before_arr=len(range_list)
        k=0
        j=1
        rearr_cycle=0
        while True:
            if(k==len(range_list)):
                break
            if((k+j)==len(range_list)):
                k=k+1
                j=1
                continue
            prev_range=range_list[k]
            curr_range=range_list[k+j]
            prev_bottomleft=prev_range[0]
            prev_topright=prev_range[1]
            curr_bottomleft=curr_range[0]
            curr_topright=curr_range[1]

            '''curr_bl_in_prev=(curr_bottomleft[0]<=prev_topright[0] and curr_bottomleft[0]>=prev_bottomleft[0]) and \
                            (curr_bottomleft[1]<=prev_topright[1] and curr_bottomleft[1]>=prev_bottomleft[1])
            curr_tr_in_prev=(curr_topright[0]<=prev_topright[0] and curr_topright[0]>=prev_bottomleft[0]) and \
                            (curr_topright[1]<=prev_topright[1] and curr_topright[1]>=prev_bottomleft[1])  
            
            prev_bl_in_curr=(prev_bottomleft[0]<=curr_topright[0] and prev_bottomleft[0]>=curr_bottomleft[0]) and \
                            (prev_bottomleft[1]<=curr_topright[1] and prev_bottomleft[1]>=curr_bottomleft[1])
            prev_tr_in_curr=(prev_topright[0]<=curr_topright[0] and prev_topright[0]>=curr_bottomleft[0]) and \
                            (prev_topright[1]<=curr_topright[1] and prev_topright[1]>=curr_bottomleft[1]) '''
            
            prev_center_x=(prev_bottomleft[0]+prev_topright[0])/2
            prev_center_y=(prev_bottomleft[1]+prev_topright[1])/2
            curr_center_x=(curr_bottomleft[0]+curr_topright[0])/2
            curr_center_y=(curr_bottomleft[1]+curr_topright[1])/2
            center_dx=abs(prev_center_x-curr_center_x)
            center_dy=abs(prev_center_y-curr_center_y)
            prev_width=abs(prev_bottomleft[0]-prev_topright[0])
            prev_height=abs(prev_bottomleft[1]-prev_topright[1])
            curr_width=abs(curr_bottomleft[0]-curr_topright[0])
            curr_height=abs(curr_bottomleft[1]-curr_topright[1])
            condition=(center_dx<=(prev_width+curr_width)/2) and (center_dy<=(prev_height+curr_height)/2)
            
            if(condition):

                new_range=[[min(curr_bottomleft[0],prev_bottomleft[0]),min(curr_bottomleft[1],prev_bottomleft[1])],\
                [max(curr_topright[0],prev_topright[0]),max(curr_topright[1],prev_topright[1])]]
                range_list[k]=new_range
                range_list.pop(k+j)
                for coord in coord_list:
                    coord_list[k].append(coord)
                #print(k)
                #print(coord_list[k])
                coord_list.pop(k+j)
                k=0
                j=1
            else:
                j=j+1
            rearr_cycle+=1
        #print(time.time()-start_time)
        #for r in range_list:
        #    print(r)






        #--------------------产生随机颜色----------------------------------------
        def randomcolor(): #随机颜色生成器
            colorArr = ['1','2','3','4','5','6','7','8','9','A','B','C','D','E','F']
            color = ""
            for i in range(6):
                color += colorArr[random.randint(0,14)]
            return "#"+color


        #------------------------绘图---------------------------------
        fig1 = plt.figure(figsize=(5,5),dpi=180)
        ax1 = fig1.add_subplot(111, aspect='equal')
        for i in range(0,len(random_list)):
            start=random_list[i][0][0]
            end=random_list[i][1][0]
            start_x=start[0]
            start_y=start[1]
            end_x=end[0]
            end_y=end[1]
            origin_x=min(start_x,end_x)
            origin_y=min(start_y,end_y)
            width=abs(start_x-end_x)
            height=abs(start_y-end_y)
            random_color=randomcolor()
            ax1.add_patch(
                patches.Rectangle(
                    (origin_x,origin_y),   # (x,y)
                    (width),          # width
                    (height),          # height
                    color=random_color,
                    fill=False
                )
            )

        for i in range(0,len(range_list)):
            start=range_list[i][0]
            end=range_list[i][1]
            start_x=start[0]
            start_y=start[1]
            end_x=end[0]
            end_y=end[1]
            origin_x=min(start_x,end_x)
            origin_y=min(start_y,end_y)
            width=abs(start_x-end_x)
            height=abs(start_y-end_y)
            random_color=randomcolor()
            ax1.add_patch(
                patches.Rectangle(
                    (origin_x,origin_y),   # (x,y)
                    (width),          # width
                    (height),          # height
                    color='k',
                    fill=False
                )
            )
        xlim(-200,400)
        ylim(-200,400)
        plt.text(100,400, '{0}\nGroup:{1}/{3}\nTime:{2}/{4}cycle'.format(range_list,len(range_list)\
                ,time.time()-start_time,len_before_arr,rearr_cycle), \
                 fontsize=7,color='k',horizontalalignment="center",verticalalignment="top")

        #plt.show()
        fig1.savefig('./random_block_pic/rect{0}.png'.format(attemp), dpi=100, bbox_inches='tight')
test_random_block_detect(30)
```
